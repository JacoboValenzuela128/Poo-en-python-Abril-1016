# Lenguaje python, crear la clase "animal" con los atributos "Nombrel" , "Raza" . Crear el contructor animal, y las funciones "ladrar()", "correr()", crear los objetos "pitbull", chihuawenio", nivel principiante, no es interactivo, proporciona tu los datos

```python
# Definición de la clase
class Animal:
    # El constructor (__init__) inicializa los atributos al crear el objeto
    def __init__(self, nombre, raza):
        self.nombre = nombre
        self.raza = raza

    # Función para simular el ladrido
    def ladrar(self):
        print(f"{self.nombre} dice: ¡Guau, guau! 🐶")

    # Función para simular la carrera
    def correr(self):
        print(f"{self.nombre} está corriendo velozmente. ⛈")

    def morder(self):
        print(f"{self.nombre} está mordiendo a la vecina. 🤦‍♂️")

# --- Creación de los objetos (Instanciación) ---

# Creamos el objeto 'pitbull'
pitbull = Animal("Zeus", "Pitbull")

# Creamos el objeto 'chihuawenio'
chihuawenio = Animal("Chispita", "Chihuahua")

# Creamos el perro callejero con datos especificos
perrocallejero = Animal("solovino", "chafa")

# --- Pruebas de funcionamiento ---

print(f"Datos del primer perro: {pitbull.nombre}, Raza: {pitbull.raza}")
pitbull.ladrar()
pitbull.correr()

print("-" * 30)

print(f"Datos del segundo perro: {chihuawenio.nombre}, Raza: {chihuawenio.raza}")
chihuawenio.ladrar()
chihuawenio.correr()

print("-" * 30)

print(f"Datos del perro callejero: {perrocallejero.nombre}, Raza: {perrocallejero.raza}")
perrocallejero.ladrar()
perrocallejero.correr()
perrocallejero.morder()

print("Programa hecho por Arath Valenzuela en ayuda de Gemini. NL:52")
```
